```python
#changing date format
start_date = datetime.strptime("1981-01", "%Y-%m")
end_date = datetime.strptime("2023-01", "%Y-%m")

date_list = pd.date_range(start_date, end_date, freq='M')
table['Date'] = date_list
table
```

```python
#checking for the distribution of percent change of earnings
table[['Total (house and land)', 'House only', 'Land only']].pct_change().describe()
```

```python
#Grouping the data by quarter
table = table.groupby(pd.Grouper(key='Date', freq='3M')).mean()
table = table.reset_index()
table.to_csv('C:/Users/ander/LightHouseNotes/final-project-tableau/Data/housing_price_index.csv', index=False)
table
```
